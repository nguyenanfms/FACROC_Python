```python
import os
import pandas as pd
import numpy as np

output_dir = '../dataset/processed'
os.makedirs(output_dir, exist_ok=True)

# 2. Sử dụng các hàm đã import để xuất dữ liệu
loaders = {
    'adult': load_adult,
    'compas': load_compas,
    'credit_card': load_credit_card,
    'german': load_german,
    'student_mat': lambda: load_student('mat'),
    'student_por': lambda: load_student('por')
}

print("Bắt đầu tiền xử lý và xuất file CSV...")
for name, load_func in loaders.items():
    try:
        X_df, P = load_func()
        
        # X_df lúc này đã là DataFrame nên copy() và gán cột sẽ thành công 100%
        df = X_df.copy()
        df['protected_attribute'] = P
        
        out_path = os.path.join(output_dir, f'{name}_processed.csv')
        df.to_csv(out_path, index=False)
        print(f" -> Thành công: {name} | Kích thước: {df.shape}")
    except Exception as e:
        print(f" -> Lỗi khi xử lý {name}: {e}")
```
